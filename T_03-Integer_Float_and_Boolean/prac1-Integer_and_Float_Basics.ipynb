```python
# Basic operations on Integer and Float data types
```

```python
# Integer Declaration

a = 10
print(a)

# Float Declaration

b = 10.5
print(b)
```

```python
# Multiple Numeric Variables

x = 5
y = 7.5
z = -3

print(x, y, z)


# Type Checking

print(type(x))
print(type(y))
print(type(z))

# Type Conversion

a = int(5.9)
b = float(10)

print(a)
print(b)
```

```python
# Arithmetic Operators

a = 10
b = 3

print(a + b)
print(a - b)
print(a * b)
print(a / b)
print(a // b)
print(a % b)
print(a ** b)

# Operator Precedence

result = 10 + 5 * 2
```

```python
# Updating Variables

x = 10

x = x + 5

print(x)


# Shorthand Operators

x = 20

x += 5
x -= 2
x *= 3
x /= 2

print(x)
```

```python
# Scientific Notation

num = 2.5e3

print(num)


# Rounding Numbers

num = 5.6789

print(round(num, 2))


# Absolute Value

print(abs(-25))


# Power Function

print(pow(2, 5))
```

```python
# Minimum & Maximum

print(min(5, 2, 8, 1))
print(max(5, 2, 8, 1))
```
