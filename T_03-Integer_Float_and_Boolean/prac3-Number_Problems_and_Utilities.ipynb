```python
# Numeric Utilities & Real Problems
```

```python
# Even or Odd Checking

num = 8

if num % 2 == 0:
    print("Even")
else:
    print("Odd")


# Positive or Negative Checking

num = -5

if num > 0:
    print("Positive")
elif num < 0:
    print("Negative")
else:
    print("Zero")







```

```python
# Largest Among Numbers

a = 10
b = 25
c = 15

print(max(a, b, c))


# Smallest Among Numbers

print(min(a, b, c))


# Swapping Variables

x = 5
y = 10

x, y = y, x

print(x, y)
```

```python
# Percentage Calculation

obtained = 450
total = 500

percentage = (obtained / total) * 100

print(percentage)


# Average Calculation

numbers = [10, 20, 30, 40]

average = sum(numbers) / len(numbers)

print(average)
```

```python
# Temperature Conversion

celsius = 37

fahrenheit = (celsius * 9/5) + 32

print(fahrenheit)


# Area Calculation

length = 10
width = 5

area = length * width

print(area)


# Simple Interest

p = 1000
r = 5
t = 2

si = (p * r * t) / 100

print(si)


# Compound Interest

p = 1000
r = 5
t = 2

ci = p * (1 + r/100) ** t

print(ci)


# Mini Calculator

a = 20
b = 4

print(a + b)
print(a - b)
print(a * b)
print(a / b)
```

```python
# Number Reversing

num = 12345

reverse = int(str(num)[::-1])

print(reverse)


# Palindrome Number Checking

num = 121

if str(num) == str(num)[::-1]:
    print("Palindrome")
else:
    print("Not Palindrome")
```

```python
# Digit Counting

num = 123456

print(len(str(num)))


# Sum of Digits

num = 1234

total = sum(int(digit) for digit in str(num))

print(total)
```

```python
# Prime Number Checking

num = 11

is_prime = True

for i in range(2, num):
    if num % i == 0:
        is_prime = False
        break

print(is_prime)


# Random Number Basics

import random

print(random.randint(1, 10))
```

```python
# Armstrong Number Checking

num = 153

digits = str(num)

power = len(digits)

result = sum(int(digit) ** power for digit in digits)

print(result == num)



# Math Library Basics

import math

print(math.sqrt(25))
print(math.ceil(4.2))
print(math.floor(4.8))
print(math.pi)
```
