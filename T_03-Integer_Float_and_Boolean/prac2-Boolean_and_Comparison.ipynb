```python
# Boolean & Comparison Operations
# ==========================================

# Boolean Values

is_true = True
is_false = False

print(is_true)
print(is_false)

# Comparison Operators

a = 5
b = 10

print(a == b)
print(a != b)
print(a < b)
print(a > b)
print(a <= b)
print(a >= b)

```

```python
# Boolean Conversion

print(bool(1))
print(bool(0))
print(bool(""))
print(bool("Python"))

# Truthy & Falsy Values

print(bool([]))
print(bool([1, 2]))
print(bool(None))
```

```python
# Logical Operators

x = True
y = False

print(x and y)
print(x or y)
print(not x)


# Boolean Expressions

age = 20

print(age >= 18)
```

```python
# Membership Comparisons

text = "Python"

print("P" in text)
print("z" not in text)


# Identity Operators

a = 10
b = 10

print(a is b)
print(a is not b)
```

```python
# Logical Operators

x = True
y = False

print(x and y)
print(x or y)
print(not x)


# Boolean Expressions

age = 20

print(age >= 18)


# Nested Logical Conditions

age = 25
salary = 50000

print(age > 18 and salary > 30000)
```
