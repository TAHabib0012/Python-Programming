```python
# Searching, Checking & Manipulation on strings
```

```python
# Declearing variables

name1 = "Hello, Peeps!"
name2 = "Enjoying Snake!!"
name3 = "Bon Appetit"
```

```python
# find and reverse find

print(name1.find("e"))    # returns the lowest index of substring 'e' in the string
print(name1.rfind("e"))   # returns the highest index of substring 'e' in the string
```

```python
# index & reverse index

print(name1.index("e"))    # returns the lowest index of substring 'e' in the string, raises ValueError if not found
print(name2.rindex("e"))   # returns the highest index of substring 'e' in the string, raises ValueError if not found
```

```python
# count, startswith & endswith

print(name1.count('e'))   # counts occurence of the mentioned substring within the string
print(name2.startswith('E'))    #checks if the string starts with the mentioned substring, prints boolean output
print(name3.endswith('n'))    #checks if the string ends with the mentioned substring, prints boolean output
```

```python
# partition & reverse partition

print(name1.partition('e'))   # partitions the string into a 3-tuple: (head, sep, tail) based on the first occurrence of the separator
print(name2.rpartition('e'))    # partitions the string into a 3-tuple: (head, sep, tail) based on the last occurrence of the separator
```

```python
# split, rsplit & join

print(name1.split('e')) # splits the string into a list of substrings using 'e' as the delimiter
print(name2.rsplit('n')) # splits the string into a list of substrings using 'n' as the delimiter, starting from the right
print(''.join(["Bon", "Appetit", "Baby", "!"])) # concatenates the elements of the list into a single string, using a space as the separator
```

```python
# removeprefix & removesuffix

print(name1.removeprefix('Hello')) # removes the specified prefix from the beginning of the string if it exists
print(name2.removesuffix('Snake')) # removes the specified suffix from the end of the string only if the string ends with that exact suffix
print(name2.removesuffix('ke!!')) # removes the specified suffix from the end of the string only if the string ends with that exact suffix
```

```python
# isdigit, isalpha, isalnum, islower, isupper, isspace & istitle

print(name1.isdigit()) # checks if all characters are digits
print(name1.isalpha()) # checks if all characters are alphabetic
print(name1.isalnum()) # checks if all characters are alphanumeric (letters or numbers)
print(name1.islower()) # checks if all cased characters are lowercase
print(name1.isupper()) # checks if all cased characters are uppercase
print(name1.isspace()) # checks if all characters are whitespace
print(name1.istitle()) # checks if the string is titlecased (e.g., words start with uppercase, others lowercase)
```
