```python
# String Methods & Transformations
```

```python
# Declering variables

name1 = "Hello, Peeps!"
name2 = "Enjoying Snake!!"
name3 = "Bon Appetit"
```

```python
# upper, lower & capitalize

print(name1.upper())
print(name2.lower())
print(name3.capitalize())
```

```python
print(name1.title()) # Converts the first letter of each word to uppercase
print(name2.swapcase()) # Swaps uppercase characters to lowercase and vice versa
print('ß'.casefold()) # Converts string to caseless matching, useful for locale-independent comparisons
print(name3.replace('Bon', 'Good')) # Replaces occurrences of a substring
```

```python
text_with_spaces = "   Hello, World!   "
print(f"Original: '{text_with_spaces}'")
print(f"Strip: '{text_with_spaces.strip()}'")     # Removes leading and trailing whitespace
print(f"LStrip: '{text_with_spaces.lstrip()}'")   # Removes leading whitespace
print(f"RStrip: '{text_with_spaces.rstrip()}'")   # Removes trailing whitespace
```

```python
print(f"Centered: '{name1.center(20, '*')}'") # Centers string within a specified width, padding with a character
print(f"Left Justified: '{name1.ljust(20, '-')}'") # Left justifies string, padding to the right
print(f"Right Justified: '{name1.rjust(20, '+')}'") # Right justifies string, padding to the left

number_str = "42"
print(f"Zero Filled: '{number_str.zfill(5)}'") # Pads numeric string with zeros on the left to a specified width
```

```python
# Chaining(converting and replacing character or substrings) methods on a substring

print(name1[0:5].upper().replace("L", "X"))  # Chaining methods on a substring
print(name2[0:6].lower().replace("P", "C"))  # Chaining methods on a substring

print(name1[7:].replace("Peeps", "Everyone").upper())  # Chaining on sliced string
print(name2[7:].replace("Programming", "Coding").lower())  # Chaining on sliced string

print(name1[0:5].upper().replace("L", "X").split("X"))  # Multiple chained methods
print(name2[0:6].lower().replace("p", "C").split("o"))  # Multiple chained methods

print(name3.upper().replace('BON', 'KHAO').title())
```
