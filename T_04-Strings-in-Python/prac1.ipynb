```python
"""
String operations examples, reorganized with short labels to make learning easier.
"""
```

```python
# Base strings used in examples
statement1 = "Hello, World!"
statement2 = "Python Programming"
```

```python
# Comparison between strings

str1 = "Hello"
str2 = "World"
print("HELLO" == "Hello")  # False (compares based on ASCII values and case sensitivity ASCII value for A is 65 and a is 97)
print("HELLO" != "hello")  
print("HELLO" < "hello")   
print("hellO" > "hello")   
```

```python
# Concatenation
print(statement1 + " " + statement2)
print(statement2 + " " + statement1)
print(statement1[0:5] + " " + statement2[0:6]) # Concatenate substrings
```

```python
# String repetition (if we do n >= 0 it will print empty string like the 2nd example)
print(statement1 * 2)
print(statement2 * 0)
```

```python
# Checking if the mentioned word is available in the string or not
print("Hello" in statement1)
print("Tik Tik" in statement1)
print("Chi chi" not in statement2)
print("Python" not in statement2)
```

```python
# Length of the strings
print(len(statement1))
print(len(statement2))
print(len(statement1 + statement2))
print(len(statement1) + len(statement2))
print(len(statement1 + statement2) * 2)
print(len(statement1) * 2 + len(statement2) * 2)
```

```python
#Escape sequence operations

print("Hello\nWorld")  # \n is a newline character
print("Hello\tWorld")  # \t is a tab character
print("Hello\\World")  # \\ is an escape sequence for a backslash. The 1st backslash allows the 2nd backslash to be printed.

# \xhh allows us to write a hexadecimal number(hh = hexadecimal). 
print("SHe is \x61 \x62itch")   #\x61 is 'a'(6*16^1 + 1 * 16^0 = 97 = a) (6*16^1 + 2 * 16^0 = 98 = b)

# \xxx (x is a digit) will allow us writting octal numbers. 
print("SHe is \141 \142itch")   #\141 is 'a'(1*8^2 + 4*8^1 + 1 * 8^0 = 49 = 1) (1*8^2 + 4*8^1 + 2 * 8^0 = 98 = 2)
```

```python
# String formatting operations

name = "Alice"
age = 30
print("My name is {} and I am {} years old.".format(name, age))
print(f"My name is {name} and I am {age} years old.")

#Using string formatter
print("My name is %s and I am %d years old." % (name, age))  # %s for string, %d for integer
print("My height is %.1f feet." % 5.9)  # %.1f for floating-point numbers
print("I'm %c student." %'a')  # %c for character
```

```python
# Converting whole string to uppercase or lowercase or capitalized
print(statement1.upper())   # Convert statement1 to uppercase
print(statement2.lower())   # Convert statement2 to lowercase
print(statement1.capitalize())  # Capitalize the first letter of statement1
print(statement2.capitalize())  # Capitalize the first letter of statement2

# Converting substrings to uppercase or lowercase
print(statement1[0:5].upper())   # Convert substring of statement1 to uppercase
print(statement2[7:18].lower())  # Convert substring of statement2 to lowercase

# Converting substrings to uppercase or lowercase from side, but printing whole strings
print(statement1[0:5].upper() + statement1[5:])   # Convert substring of statement1 to uppercase
print(statement2[:7] + statement2[7:18].lower() + statement2[18:])  # Convert substring of statement2 to lowercase

# Converting substrings to uppercase or lowercase from middle, but printing whole strings
print(statement1[:5] + statement1[5:10].upper() + statement1[10:])  # Convert substring of statement1 to uppercase
print(statement2[:7] + statement2[7:18].lower() + statement2[18:])  # Convert substring of statement2 to lowercase

# Converting substrings to uppercase or lowercase from middle, but printing whole strings
print(statement1[:5] + statement1[5:10].upper() + statement1[10:])  # Convert substring of statement1 to uppercase
print(statement2[:7] + statement2[7:18].lower() + statement2[18:])  # Convert substring of statement2 to lowercase
```

```python
# Replacing substrings from whole string  
print(statement1.replace("Hello", "Hi"))
print(statement2.replace("Python", "Java"))
```

```python
# Slicing  the string then replacing the substrings from sliced string
print(statement1[7:].replace("World", "Everyone"))
print(statement2[7:].replace("Programming", "Coding"))
```

```python
# Splitting  the string to look like an array and joining different strings to make one string
print(statement1.split(","))  # Split into list by comma
print(statement2.split(" "))  # Split into list by space
print(statement2.join(["Let's", "learn", "Python"]))  # Join list into string
print(statement2.split("o"))  # Split into list by 'o'

# reverse split
print(statement2.rsplit("n"))  # Split into list by 'o' from the right
```

```python
# Searching, counting, checks, indexing

print(statement1.startswith("Hello"))
print(statement2.endswith("Programming"))
print(statement1.find("o")) # Find the first occurrence of "o"
print(statement1.rfind("o")) # Find the last occurrence of "o" 
print(statement2.count("o")) # Count the occurrences of "o"
print(statement2.isdigit())  # Check if string contains only digits
print(statement1.index("o")) # Find the first occurrence of "o"
print(statement1.rindex("o")) # Find the last occurrence of "o"
```

```python
# stripping
statement3 = "   Hello World   "


print(statement3.strip())  # Remove whitespace from both ends
print(statement3.lstrip())  # Remove whitespace from the left end
print(statement3.rstrip())  # Remove whitespace from the right end
print("###Hello World###".strip('#'))  # Remove '#' from both ends, can do with any character
print(statement1.strip('oldH'))  
# here each character in 'oldH' is thought as a separate character to remove, 
# so if any of those characters are present in the string at beginning or end, they will be removed
```

```python
# Chaining(converting and replacing character or substrings) methods on a substring
print(statement1[0:5].upper().replace("L", "X"))  # Chaining methods on a substring
print(statement2[0:6].lower().replace("P", "C"))  # Chaining methods on a substring

print(statement1[7:].replace("World", "Everyone").upper())  # Chaining on sliced string
print(statement2[7:].replace("Programming", "Coding").lower())  # Chaining on sliced string

print(statement1[0:5].upper().replace("L", "X").split("X"))  # Multiple chained methods
print(statement2[0:6].lower().replace("p", "C").split("o"))  # Multiple chained methods
```

```python
# Implicit type conversion
num1 = 5
num2 = 10.25
print("The sum is: " + str(num1 + num2))
# division
print(10 / 2) #  this division will have float ans
print(10 // 2) # this division will have int ans


print("The sum is: " + (num1 + num2)) # this will give  error as 1st portion is string and 2nd is union of int and float

```

```python
# Explicit type conversion
num1 = 5
num2 = 10.25
s = "hey"
num3 =  int(num2)  # Convert num2 to integer
num4 = str(num1)  # Convert num1 to string
num5 = float(s)  # Convert s to float, but sentence/character is not possible to convert

print(num3)  
print(num4)  
print(num5)  
```
