```python
# Multiple string variables
name1 = "Hello, World!"
name2 = "Python Programming"
print(name1)
print(name2)
```

```python
# String indexing

print(name1[0])
print(name2[9])

# Negative indexing

print(name1[-3])
print(name2[-5])
```

```python
# String length (len)

print(len(name1))
print(len(name2))
```

```python
# String slicing

print(name1[0:5])
print(name2[8:18])
print(name1[7:])
print(name2[:9])
```

```python
# String Immutability
# In Python, strings are immutable, meaning once a string is created, its content cannot be changed.
# Any operation that appears to modify a string, like concatenation or slicing, actually creates a new string.

my_string = "hello"
print(f"Original string: {my_string}")

# Attempting to change a character will result in a TypeError
try:
    my_string[0] = 'H' # This line will raise an error
except TypeError as e:
    print(f"\nError when trying to modify a character: {e}")

# To 'change' a string, you must create a new one
new_string = "H" + my_string[1:]
print(f"New string after 'modification': {new_string}")

# Demonstrating that the original string is unchanged
print(f"Original string (still): {my_string}")

# You can also assign the new string back to the original variable name
my_string = my_string + " world"
print(f"String after concatenation (a new string assigned to my_string): {my_string}")
```

```python
# Membership operators, String repetition & String concatenation

print('hello' in name1)
print('hello' not in name2)

print(name1 * 3)
print(name2 * 0)

print(name1 + name2)
print(name2 + ' ' + '&' + ' ' + name1)
```

```python
#String comparison & ASCII comparison
str1 = "apple"
str2 = "APPLE"

print(str1 == str2)
print(str1 != str2)
print(str1 > str2)
print(str1 < str2)

print(ord('a'))
print(ord('A'))
```

```python
# Multiline strings
multi_line_string = """This is a string
that spans multiple
lines."""
print(multi_line_string)

# Raw strings
# Raw strings treat backslashes as literal characters, useful for file paths or regular expressions.
file_path = r"C:Users\Documents\new_folder\file.txt"
print(f"File path (raw string): {file_path}")

# Compare with a regular string where backslashes are escape characters
regular_string_path = "C:Users\Documents\new_folder\file.txt"
print(f"File path (regular string): {regular_string_path}")
```

```python
# Escape Sequences
# Escape sequences allow you to include special characters in strings.
# They start with a backslash (\).

print("1. Newline: This is on the first line.\nThis is on the second line.")
print("2. Tab: Item1\tItem2\tItem3")
print("3. Backslash: To print a literal backslash, use double backslash: \\")
print("4. Single Quote: It\'s a beautiful day.")
print("5. Double Quote: He said, \"Hello!\"")

# Some less common ones:
print("6. Backspace: Word\b\b\bAnother") # Backspace (moves cursor back)
print("7. Carriage Return: Line 1\rLine 2") # Carriage Return (moves cursor to start of line)
print("8. Form Feed: Page 1\fPage 2") # Form Feed (often treated as a page break)
print("9. Hexadecimal Value: \x41 is 'A'") # Character by hex value
print("10. Octal Value: \101 is 'A'") # Character by octal value

# Example of a path with escape sequences vs. a raw string:
print("\nPath (with escapes): C:\\Users\\Documents\\new_folder\\file.txt")
print(r"Path (raw string): C:\Users\Documents\new_folder\file.txt")
```
