```python
# Palindrome checking

name1 = "ABBA"
name2 = "AbBA"

print(name1 == name1[::-1])
print(name2 == name2[::-1])

if name1 == name1[::-1]:
    print("Palindrome")
else:
    print("Not a palindrome")
```

```python
# Word counting

name1 = "I'm Snoop Dogg"

print(name1.split())
print(len(name1.split()))
```

```python
# Removing punctuation
import string

name1 = "I'm Snoop-Dogg."
name2 = ""


for ch in name1:
  if ch not in string.punctuation:
    name2 += ch

print(name2)
```

```python
# Removing extra spaces

name1 = "     I'm     Snoop Dogg   "

print(name1.split())
print(" ".join(name1.split()))
```

```python
# Character frequency counting
# Approach - 1

# Define the input string for character frequency analysis.
name1 = "I'm Snoop Dogg"

# Initialize an empty dictionary to store character frequencies.
freq = {}

# Iterate over each character in the string 'name1'.
for ch in name1:
  # For each character, increment its count in the 'freq' dictionary.
  # If the character is not yet in the dictionary, add it with a count of 1.
  freq[ch] = freq.get(ch, 0) + 1


# Print the resulting dictionary containing character frequencies.
print(freq)


# Approach - 2

# Character frequency counting using collections.Counter
import collections

# Use collections.Counter to count character frequencies.
# Counter is a subclass of dictionary for counting hashable objects.
freq_counter = collections.Counter(name1)

# Print the resulting Counter object.
print(freq_counter)
```

```python
# String cleaning
import re
import string

# Original string with various issues
raw_text = "   Hello,  World!  This is a TEST string with Punctuation & Numbers 123.   "
print(f"Original: '{raw_text}'")

# 1. Remove leading/trailing whitespace
stripped_text = raw_text.strip()
print(f"1. Stripped: '{stripped_text}'")

# 2. Convert to lowercase
lowercase_text = stripped_text.lower()
print(f"2. Lowercase: '{lowercase_text}'")

# 3. Remove punctuation
# Using string.punctuation and a list comprehension
no_punctuation_text = "".join([char for char in lowercase_text if char not in string.punctuation])
print(f"3. No punctuation: '{no_punctuation_text}'")

# 4. Replace multiple spaces with a single space
# Using split() and join() to replace one or more spaces with a single space, without regex.
cleaned_text = " ".join(no_punctuation_text.split())
print(f"4. Cleaned (single spaces): '{cleaned_text}'")
```

```python
#  translate

text = "hello"

table = str.maketrans("helo", "HELO")

result = text.translate(table)

print(result)
```

```python
# maketranslate

text = "apple"

translation_table = str.maketrans({
    "a": "@",
    "e": "3"
})

result = text.translate(translation_table)

print(result)
```
