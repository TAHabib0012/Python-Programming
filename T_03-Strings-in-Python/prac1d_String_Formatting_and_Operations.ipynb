```python
# Declearing variables

name1 = "Hello, Peeps!"
name2 = "Enjoying Snake!!"
name3 = "Bon Appetit"
```

```python
# Example of arithmetic/string conversion issues

str_num1 = '10'
str_num2 = '5'
int_num = 2

# This will raise a TypeError: can only concatenate str (not "int") to str
# print(str_num1 + int_num)

# This will concatenate the strings, not add the numbers
print(f"'10' + '5' (string concatenation): {str_num1 + str_num2}")

# Correct way to perform arithmetic: convert strings to numbers first
result_add = int(str_num1) + int(str_num2)
result_multiply = int(str_num1) * int_num
result_float_add = float(str_num1) + float(str_num2)

print(f"10 + 5 (after conversion to int): {result_add}")
print(f"10 * 2 (after conversion to int): {result_multiply}")
print(f"10.0 + 5.0 (after conversion to float): {result_float_add}")

# Attempting to convert a non-numeric string will raise a ValueError
non_numeric_str = 'hello'
# int(non_numeric_str) # Uncomment to see ValueError
```

```python
# Example of ord() function.
#The ord() function in Python returns the Unicode code point of a single character string

char1 = 'A'
char2 = ' ' # space
char3 = '€'

# Get the Unicode code point for each character
code_point1 = ord(char1)
code_point2 = ord(char2)
code_point3 = ord(char3)

print(f"The Unicode code point for '{char1}' is: {code_point1}")
print(f"The Unicode code point for '{char2}' is: {code_point2}")
print(f"The Unicode code point for '{char3}' is: {code_point3}")

# ord() only accepts a single character string
# ord("AB") # Uncomment to see TypeError
```

```python
# Example of chr() function
'''The chr() function in Python is the inverse of ord(). It takes an integer
representing a Unicode code point and returns the corresponding character string.'''

code_point1 = 65  # Unicode for 'A'
code_point2 = 97  # Unicode for 'a'
code_point3 = 8364 # Unicode for '€'

# Convert code points back to characters
char_from_code1 = chr(code_point1)
char_from_code2 = chr(code_point2)
char_from_code3 = chr(code_point3)

print(f"The character for code point {code_point1} is: {char_from_code1}")
print(f"The character for code point {code_point2} is: {char_from_code2}")
print(f"The character for code point {code_point3} is: {char_from_code3}")

# chr() expects an integer within the valid Unicode range
# chr(1114112) # Uncomment to see ValueError for out of range
```

```python
# Example of min() function with strings and characters

string_data = "hello world"
string_compare_1 = "apple"
string_compare_2 = "banana"
string_compare_3 = "zebra"

# Find the minimum character within a string (based on Unicode value)
# For 'hello world', ' ' (space) has a lower Unicode value than 'h'
min_char_in_string = min(string_data)
print(f"Minimum character in '{string_data}': '{min_char_in_string}'")

# Find the minimum among multiple string arguments (lexicographical order)
min_of_strings = min(string_compare_1, string_compare_2, string_compare_3)
print(f"Minimum among ('{string_compare_1}', '{string_compare_2}', '{string_compare_3}'): '{min_of_strings}'")

# Find the minimum of single characters
min_of_chars = min('X', 'A', 'Z')
print(f"Minimum of ('X', 'A', 'Z'): '{min_of_chars}'")

# Using a key to find the string with the minimum length
min_len_string = min(string_compare_1, string_compare_2, string_compare_3, key=len)
print(f"String with minimum length among ('{string_compare_1}', '{string_compare_2}', '{string_compare_3}'): '{min_len_string}'")
```

```python
# Example of max() function with strings and characters

string_data = "hello world"
string_compare_1 = "apple"
string_compare_2 = "banana"
string_compare_3 = "zebra"

# Find the maximum character within a string (based on Unicode value)
max_char_in_string = max(string_data)
print(f"Maximum character in '{string_data}': '{max_char_in_string}'")

# Find the maximum among multiple string arguments (lexicographical order)
max_of_strings = max(string_compare_1, string_compare_2, string_compare_3)
print(f"Maximum among ('{string_compare_1}', '{string_compare_2}', '{string_compare_3}'): '{max_of_strings}'")

# Find the maximum of single characters
max_of_chars = max('X', 'A', 'Z')
print(f"Maximum of ('X', 'A', 'Z'): '{max_of_chars}'")

# Using a key to find the string with the maximum length
max_len_string = max(string_compare_1, string_compare_2, string_compare_3, key=len)
print(f"String with maximum length among ('{string_compare_1}', '{string_compare_2}', '{string_compare_3}'): '{max_len_string}'")
```

```python
# Example of sorted() function with a string

my_string = "python programming"

# Sorting a string produces a list of characters
sorted_chars_list_asc = sorted(my_string)
print(f"Original string: '{my_string}'")
print(f"Sorted characters (as list, ascending): {sorted_chars_list_asc}")

# To get a sorted string, join the characters back together
sorted_string_asc = ''.join(sorted_chars_list_asc)
print(f"Sorted string (ascending): '{sorted_string_asc}'")

# Sort in descending order
sorted_chars_list_desc = sorted(my_string, reverse=True)
print(f"Sorted characters (as list, descending): {sorted_chars_list_desc}")

sorted_string_desc = ''.join(sorted_chars_list_desc)
print(f"Sorted string (descending): '{sorted_string_desc}'")

# Sorting a string with a key function (e.g., case-insensitive sort)
case_mix_string = "BananaAppleOrange"
sorted_case_insensitive_list = sorted(case_mix_string, key=str.lower)
sorted_case_insensitive_string = ''.join(sorted_case_insensitive_list)
print(f"Original string: '{case_mix_string}'")
print(f"Sorted string (case-insensitive): '{sorted_case_insensitive_string}'")
```

```python
# Example of enumerate() function with a string

my_string = 'Python'

# Iterate through a string with index and character
print("Iterating through 'Python' with index and character:")
for index, char in enumerate(my_string):
    print(f"Index: {index}, Character: '{char}'")

# Start the counter from a different number (e.g., 1)
print("\nIterating with start=1:")
for i, char in enumerate(my_string, start=1):
    print(f"{i}. '{char}'")

# If you need a list of (index, character) tuples, you can convert it
# Although the user specified 'only strings', this shows the output structure.
enumerated_list = list(enumerate(my_string))
print(f"\nEnumerate object converted to list of tuples (index, char): {enumerated_list}")
```

```python
# Example of String Reversing using string-only operations

my_string = "hello world"

# Method: Using slicing [start:end:step]
# A step of -1 reverses the string
reversed_string_slice = my_string[::-1]

print(f"Original string: '{my_string}'")
print(f"Reversed using slicing: '{reversed_string_slice}'")

# Another way to think about reversing specific parts using slicing
part_to_reverse = my_string[6:11] # 'world'
reversed_part = part_to_reverse[::-1] # 'dlrow'
print(f"\nOriginal part: '{part_to_reverse}'")
print(f"Reversed part: '{reversed_part}'")
```

```python
# Example of String Sorting (character-wise)

my_string = "python programming"
another_string = "bacde"

# Sort a string alphabetically (ascending) character-wise
# `sorted(my_string)` produces a list of characters, then ''.join() converts it back to a string.
sorted_string_asc = ''.join(sorted(my_string))
print(f"Original string: '{my_string}'")
print(f"Sorted characters (as a new string, ascending): '{sorted_string_asc}'")

# Sort a string alphabetically (descending) character-wise
sorted_string_desc = ''.join(sorted(another_string, reverse=True))
print(f"Original string: '{another_string}'")
print(f"Sorted characters (as a new string, descending): '{sorted_string_desc}'")

# Sorting based on a custom key (e.g., case-insensitive sort)
case_mix_string = "BananaAppleOrange"
sorted_case_insensitive = ''.join(sorted(case_mix_string, key=str.lower))
print(f"Original string: '{case_mix_string}'")
print(f"Sorted characters (as a new string, case-insensitive): '{sorted_case_insensitive}'")

# Note: The original string remains unchanged due to string immutability.
print(f"Original string after sorting attempts: '{my_string}'")
```

```python
# Example of String Iteration

my_string = "Colab"

# Iterate through each character in the string
print("Iterating through 'Colab':")
for char in my_string:
    print(char)

# Iterate with index using enumerate
print("\nIterating with index and character:")
for i, char in enumerate(my_string):
    print(f"Character at index {i}: {char}")

# Iterate with a while loop (less common but possible)
print("\nIterating with a while loop:")
index = 0
while index < len(my_string):
    print(my_string[index])
    index += 1
```

```python
# Example of Encoding Basics (encode())

my_string_ascii = "Hello"
my_string_unicode = "你好 world €"

# Encode using UTF-8 (common encoding)
encoded_utf8 = my_string_ascii.encode('utf-8')
print(f"Original string (ASCII): '{my_string_ascii}'")
print(f"Encoded to UTF-8: {encoded_utf8} (Type: {type(encoded_utf8)})\n")

# Encode a string with non-ASCII characters to UTF-8
encoded_unicode_utf8 = my_string_unicode.encode('utf-8')
print(f"Original string (Unicode): '{my_string_unicode}'")
print(f"Encoded to UTF-8: {encoded_unicode_utf8}\n")

# Encode using ASCII (will raise error for non-ASCII characters)
# The 'errors' parameter can handle this:
# 'strict' (default): raises UnicodeEncodeError
# 'ignore': ignores unencodable characters
# 'replace': replaces unencodable characters with a placeholder

# This will likely cause an error if '你好' is present
try:
    encoded_ascii_strict = my_string_unicode.encode('ascii')
except UnicodeEncodeError as e:
    print(f"Error encoding to ASCII (strict): {e}\n")

# Encode to ASCII, ignoring errors
encoded_ascii_ignore = my_string_unicode.encode('ascii', errors='ignore')
print(f"Encoded to ASCII (ignore errors): {encoded_ascii_ignore}\n")

# Encode to ASCII, replacing errors
encoded_ascii_replace = my_string_unicode.encode('ascii', errors='replace')
print(f"Encoded to ASCII (replace errors): {encoded_ascii_replace}\n")

# Example of encoding to a different encoding like UTF-16
encoded_utf16 = my_string_unicode.encode('utf-16')
print(f"Encoded to UTF-16: {encoded_utf16}")
```
